```
gpt2.download_gpt2(model_name="355M")
```

```
gpt2.mount_gdrive()
```

```
file_name = 'full_tas.txt'
```

```
gpt2.copy_file_from_gdrive(file_name)
```

```
sess = gpt2.start_tf_sess()

gpt2.finetune(sess,
              dataset=file_name,
              model_name='355M',
              steps=1000,
              restore_from='fresh',
              run_name='run3_StarTrek_tas',
              print_every=10,
              sample_every=200,
              save_every=500
              )
```

```
gpt2.copy_checkpoint_to_gdrive(run_name='run3_StarTrek_tas')
```

```
text_startrek_tas = gpt2.generate(sess,
                              run_name='run3_StarTrek_tas',
                              length=500, prefix=' SWITCH ',
                              temperature=0.9, nsamples=10, batch_size=10, 
                              return_as_list=True)
```

```
with open('generated_text_startrek_tas.txt', 'w', encoding='utf-8') as f:
  f.writelines(text_startrek_tas)

!cp generated_text_startrek_tas.txt /content/drive/MyDrive
```
